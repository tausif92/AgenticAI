### E2E Agentic AI Project
This project is built and deployed using streamlit.  
It will have majorly 3 usecases:
1. Basic chatbot
2. Chatbot with tools
3. AI news

When designing the E2E project, there are few important components :
1. Nodes
2. Graph
3. LLMs (the functionality of nodes)
4. State (graph state)
5. UI (front-end)

These components can be put as folders in the project.
